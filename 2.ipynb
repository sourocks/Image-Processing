```python
import cv2
import numpy as np

def nill(x):
    pass

img=np.zeros((500,500,3),np.uint8)*255
cv2.namedWindow("colour")
cv2.createTrackbar("R","colour",0,255,nill)
cv2.createTrackbar("G","colour",0,255,nill)
cv2.createTrackbar("B","colour",0,255,nill)

while True:
    cv2.imshow("window",img)
    if cv2.waitKey(1) & 0xff == ord("p"):
        break
    r=cv2.getTrackbarPos("R","colour")
    g=cv2.getTrackbarPos("G","colour")
    b=cv2.getTrackbarPos("B","colour")

    img[:]=[b,g,r]

cv2.destroyAllWindows()
```

#### Histograms Equalization in OpenCV

Histogram equalization is a method in image processing of contrast adjustment using the
image's histogram.
This method usually increases the global contrast of many images, especially when the
usable data of the image is represented by close contrast values.

```python
import cv2
import numpy as np

# Initialize global variables
drawing = False  # True if the mouse is pressed
current_color = (0, 0, 255)  # Default color is red
color_palette = [(0, 0, 255), (0, 255, 0), (255, 0, 0), (0, 255, 255), (255, 255, 0), (255, 0, 255), (0, 0, 0)]  # Red, Green, Blue, Yellow, Cyan, Magenta, Black

# Create a blank whiteboard
img = np.ones((600, 800, 3), np.uint8) * 255

# Function to draw the color palette
def draw_color_palette():
    for i, color in enumerate(color_palette):
        cv2.rectangle(img, (0, i * 100), (50, (i + 1) * 100), color, -1)

draw_color_palette()  # Draw palette initially

# Mouse callback function
def draw(event, x, y, flags, param):
    global drawing, current_color
    
    # Check if clicked on color palette
    if event == cv2.EVENT_LBUTTONDOWN:
        if x <= 50:  # Palette is in the first 50 pixels
            palette_index = y // 100  # Determine which color was clicked
            if palette_index < len(color_palette):
                current_color = color_palette[palette_index]
        else:
            drawing = True
            cv2.circle(img, (x, y), 5, current_color, -1)
    
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            cv2.circle(img, (x, y), 3, current_color, -1)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False

    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img, (x, y), (x + 20, y + 20), current_color, 2)

    elif event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.line(img, (x, y), (x + 50, y + 50), current_color, 2)

# Create a window and set the mouse callback function
cv2.namedWindow("Whiteboard")
cv2.setMouseCallback("Whiteboard", draw)

while True:
    cv2.imshow("Whiteboard", img)
    if cv2.waitKey(25) & 0xff == ord("p"):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Initialize global variables
drawing = False  # True if the mouse is pressed
ix, iy = -1, -1  # Initial coordinates for drawing
current_color = (0, 0, 255)  # Default drawing color (red)

# Color palette coordinates
color_palette = [(0, 0, 255), (0, 255, 0), (255, 0, 0), (0, 255, 255), (255, 255, 0), (255, 0, 255)]
palette_height = 50

def draw(event, x, y, flags, param):
    global ix, iy, drawing, current_color
    
    # Drawing a circle or rectangle or line based on mouse events
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            cv2.line(img, (ix, iy), (x, y), current_color, 4)  # Draw line continuously
            ix, iy = x, y  # Update initial coordinates to the current ones
    
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img, (x, y), (x + 20, y + 20), (255, 0, 0), 4)  # Draw rectangle on right-click

    elif event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.line(img, (x, y), (x + 50, y + 50), (250, 0, 0), 4)  # Draw a line on double click
    
    # Check if a color on the palette is clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        if y <= palette_height:
            color_index = x // 100  # Each color block is 100 pixels wide
            if color_index < len(color_palette):
                current_color = color_palette[color_index]  # Change the current drawing color

# Create a whiteboard window
cv2.namedWindow("Whiteboard")

# Create a white image for the whiteboard
img = np.ones((600, 700, 3), np.uint8) * 255

# Draw color palette on top
for i, color in enumerate(color_palette):
    cv2.rectangle(img, (i * 100, 0), ((i + 1) * 100, palette_height), color, -1)

cv2.setMouseCallback("Whiteboard", draw)

# Main loop to display the whiteboard
while True:
    cv2.imshow("Whiteboard", img)
    if cv2.waitKey(25) & 0xff == ord("p"):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a whiteboard
whiteboard_size = (800, 1000, 3)
img = np.ones(whiteboard_size, np.uint8) * 255
img_copy = img.copy()  # Copy of the original image for undo/redo

# Initialize history for undo/redo
history = [img.copy()]
redo_stack = []

# Create a window for the whiteboard
cv2.namedWindow("Whiteboard")

# Color picker size
color_picker_size = 150
color_picker = np.zeros((color_picker_size, color_picker_size, 3), np.uint8)

# Fill the circular color picker with colors
for i in range(color_picker.shape[0]):
    for j in range(color_picker.shape[1]):
        dx, dy = j - color_picker_size // 2, i - color_picker_size // 2
        radius = np.sqrt(dx**2 + dy**2)

        if radius < color_picker_size // 2:  # Ensure it's within the circle
            if radius < 20:  # Add a white region in the center
                color_picker[i, j] = [255, 255, 255]  # White color
            else:
                angle = np.arctan2(dy, dx)
                r = int(127.5 * (1 + np.cos(angle)))
                g = int(127.5 * (1 + np.cos(angle + 2 * np.pi / 3)))
                b = int(127.5 * (1 + np.cos(angle + 4 * np.pi / 3)))
                color_picker[i, j] = [b, g, r]

# Variables to track drawing state
drawing = False  # True when the left mouse button is pressed
erasing = False  # True when the right mouse button is pressed
ix, iy = -1, -1  # Initial mouse coordinates
current_color = (0, 0, 0)  # Default color (black)

# Mouse callback function for the whiteboard
def draw(event, x, y, flags, param):
    global ix, iy, drawing, erasing, img, current_color, history, redo_stack

    # Left mouse button pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        # Check if the click is within one of the buttons or color picker
        if 10 <= x <= 10 + color_picker_size and 10 <= y <= 10 + color_picker_size:  # Color picker
            pick_color(x, y)
        elif 10 + color_picker_size + 10 <= x <= 90 + color_picker_size and 10 <= y <= 40:  # Undo button
            undo()
        elif 100 + color_picker_size <= x <= 180 + color_picker_size and 10 <= y <= 40:  # Redo button
            redo()
        elif 190 + color_picker_size <= x <= 290 + color_picker_size and 10 <= y <= 40:  # Clear All button
            clear_all()
        else:
            drawing = True
            ix, iy = x, y  # Start drawing from this point
            if len(history) >= 10:
                history.pop(0)  # Maintain only the last 10 actions
            history.append(img.copy())
            redo_stack = []  # Clear redo stack
            # Draw a dot on single click
            cv2.circle(img, (x, y), 2, current_color, -1)

    # Left mouse button released
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False

    # Right mouse button pressed
    elif event == cv2.EVENT_RBUTTONDOWN:
        erasing = True

    # Right mouse button released
    elif event == cv2.EVENT_RBUTTONUP:
        erasing = False

    # Mouse movement
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            if ix != -1 and iy != -1:
                cv2.line(img, (ix, iy), (x, y), current_color, 2)  # Draw a line from the last point to the current point
                ix, iy = x, y  # Update the last point to the current point
        elif erasing:
            cv2.rectangle(img, (x, y), (x + 20, y + 20), (255, 255, 255), -1)  # Draw a white rectangle to erase

# Mouse callback function for the color picker
def pick_color(x, y):
    global current_color
    # Calculate color picker coordinates relative to whiteboard
    picker_x = x - 10
    picker_y = y - 10
    if 0 <= picker_x < color_picker_size and 0 <= picker_y < color_picker_size:
        current_color = color_picker[picker_y, picker_x].tolist()  # Update the current color

# Function to undo the last action
def undo():
    global history, img, redo_stack
    if len(history) > 1:
        redo_stack.append(history.pop())  # Move last state to redo stack
        img = history[-1].copy()  # Restore last state

# Function to redo the last undone action
def redo():
    global redo_stack, img, history
    if redo_stack:
        img = redo_stack.pop().copy()  # Restore from redo stack
        if len(history) >= 10:
            history.pop(0)  # Maintain only the last 10 actions
        history.append(img.copy())  # Add to history

# Function to clear all drawings
def clear_all():
    global img, history, redo_stack
    img = np.ones(whiteboard_size, np.uint8) * 255
    history = [img.copy()]  # Reset history
    redo_stack = []  # Clear redo stack

# Draw buttons and color picker on the whiteboard
def draw_elements():
    global img
    # Color picker
    img[10:10 + color_picker_size, 10:10 + color_picker_size] = color_picker

    # Undo button
    cv2.rectangle(img, (10 + color_picker_size + 10, 10), (90 + color_picker_size + 10, 40), (200, 200, 200), -1)
    cv2.putText(img, "Undo", (20 + color_picker_size + 10, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    # Redo button
    cv2.rectangle(img, (100 + color_picker_size, 10), (180 + color_picker_size, 40), (200, 200, 200), -1)
    cv2.putText(img, "Redo", (110 + color_picker_size, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    # Clear All button
    cv2.rectangle(img, (190 + color_picker_size, 10), (290 + color_picker_size, 40), (200, 200, 200), -1)
    cv2.putText(img, "Clear All", (200 + color_picker_size, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

# Set the mouse callback functions
cv2.setMouseCallback("Whiteboard", draw)

# Main loop
while True:
    draw_elements()
    cv2.imshow("Whiteboard", img)

    key = cv2.waitKey(1) & 0xff
    if key == ord("p"):  # Exit on 'p'
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a whiteboard
whiteboard_size = (800, 1000, 3)
img = np.ones(whiteboard_size, np.uint8) * 255
img_copy = img.copy()  # Copy of the original image for undo/redo

# Initialize history for undo/redo
history = [img.copy()]
redo_stack = []

# Create a window for the whiteboard
cv2.namedWindow("Whiteboard")

# Color picker size
color_picker_width = 30
color_picker_height = 200
color_picker = np.zeros((color_picker_height, color_picker_width, 3), np.uint8)

# Gradient sections

n_colors = 10  # Number of gradients
sections = color_picker_height // n_colors

# Extended Gradient List
gradients = [
    (np.array([255, 255, 255]), np.array([0, 0, 0])),   # White to Black
    (np.array([0, 0, 0]), np.array([0, 0, 255])),       # Black to Blue
    (np.array([0, 0, 255]), np.array([0, 255, 255])),    # Blue to Cyan
    (np.array([0, 255, 255]), np.array([0, 255, 0])),    # Cyan to Green
    (np.array([0, 255, 0]), np.array([255, 255, 0])),    # Green to Yellow
    (np.array([255, 255, 0]), np.array([255, 0, 0])),    # Yellow to Red
    (np.array([255, 0, 0]), np.array([255, 0, 255])),    # Red to Magenta
    (np.array([255, 0, 255]), np.array([0, 0, 0])),      # Magenta to Black
    (np.array([0, 0, 0]), np.array([255, 255, 255])),    # Black to White
    (np.array([255, 255, 255]), np.array([0, 0, 0]))     # White to Black
]


# Generate the color picker gradient
for i in range(color_picker_height):
    grad_index = min(i // sections, len(gradients) - 1)
    start_color, end_color = gradients[grad_index]
    t = (i % sections) / sections
    color = (1 - t) * start_color + t * end_color
    color_picker[i, :] = color.astype(np.uint8)

# Variables to track drawing state
drawing = False  # True when the left mouse button is pressed
erasing = False  # True when the right mouse button is pressed
ix, iy = -1, -1  # Initial mouse coordinates
current_color = (0, 0, 0)  # Default color (black)

# Mouse callback function for the whiteboard
def draw(event, x, y, flags, param):
    global ix, iy, drawing, erasing, img, current_color, history, redo_stack

    # Left mouse button pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        # Check if the click is within one of the buttons or color picker
        if 10 <= x <= 10 + color_picker_width and 10 <= y <= 10 + color_picker_height:  # Color picker
            pick_color(x, y)
        elif 10 + color_picker_width + 10 <= x <= 90 + color_picker_width and 10 <= y <= 40:  # Undo button
            undo()
        elif 100 + color_picker_width <= x <= 180 + color_picker_width and 10 <= y <= 40:  # Redo button
            redo()
        elif 190 + color_picker_width <= x <= 290 + color_picker_width and 10 <= y <= 40:  # Clear All button
            clear_all()
        else:
            drawing = True
            ix, iy = x, y  # Start drawing from this point
            if len(history) >= 10:
                history.pop(0)  # Maintain only the last 10 actions
            history.append(img.copy())
            redo_stack = []  # Clear redo stack
            # Draw a dot on single click
            cv2.circle(img, (x, y), 2, current_color, -1)

    # Left mouse button released
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False

    # Right mouse button pressed
    elif event == cv2.EVENT_RBUTTONDOWN:
        erasing = True

    # Right mouse button released
    elif event == cv2.EVENT_RBUTTONUP:
        erasing = False

    # Mouse movement
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            if ix != -1 and iy != -1:
                cv2.line(img, (ix, iy), (x, y), current_color, 2)  # Draw a line from the last point to the current point
                ix, iy = x, y  # Update the last point to the current point
        elif erasing:
            cv2.rectangle(img, (x, y), (x + 25, y + 25), (255, 255, 255), -1)  # Draw a white rectangle to erase

# Mouse callback function for the color picker
def pick_color(x, y):
    global current_color
    # Calculate color picker coordinates relative to whiteboard
    picker_x = x - 10
    picker_y = y - 10
    if 0 <= picker_x < color_picker_width and 0 <= picker_y < color_picker_height:
        current_color = color_picker[picker_y, picker_x].tolist()  # Update the current color

# Function to undo the last action
def undo():
    global history, img, redo_stack
    if len(history) > 1:
        redo_stack.append(history.pop())  # Move last state to redo stack
        img = history[-1].copy()  # Restore last state

# Function to redo the last undone action
def redo():
    global redo_stack, img, history
    if redo_stack:
        img = redo_stack.pop().copy()  # Restore from redo stack
        if len(history) >= 10:
            history.pop(0)  # Maintain only the last 10 actions
        history.append(img.copy())  # Add to history

# Function to clear all drawings
def clear_all():
    global img, history, redo_stack
    img = np.ones(whiteboard_size, np.uint8) * 255
    history = [img.copy()]  # Reset history
    redo_stack = []  # Clear redo stack

# Draw buttons and color picker on the whiteboard
def draw_elements():
    global img
    # Color picker
    img[10:10 + color_picker_height, 10:10 + color_picker_width] = color_picker

    # Undo button
    cv2.rectangle(img, (10 + color_picker_width + 10, 10), (90 + color_picker_width + 10, 40), (200, 200, 200), -1)
    cv2.putText(img, "Undo", (20 + color_picker_width + 10, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    # Redo button
    cv2.rectangle(img, (100 + color_picker_width, 10), (180 + color_picker_width, 40), (200, 200, 200), -1)
    cv2.putText(img, "Redo", (110 + color_picker_width, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    # Clear All button
    cv2.rectangle(img, (190 + color_picker_width, 10), (290 + color_picker_width, 40), (200, 200, 200), -1)
    cv2.putText(img, "Clear All", (200 + color_picker_width, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

# Set the mouse callback functions
cv2.setMouseCallback("Whiteboard", draw)

# Main loop
while True:
    draw_elements()
    cv2.imshow("Whiteboard", img)

    key = cv2.waitKey(1) & 0xff
    if key == ord("p"):  # Exit on 'p'
        break

cv2.destroyAllWindows()
```
